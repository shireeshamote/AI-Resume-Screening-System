```python
#Read all PDF resumes automatically
import os
import pdfplumber

resume_texts = {}
folder_path = "resumes"

for file in os.listdir(folder_path):

    if file.endswith(".pdf"):

        pdf_path = os.path.join(folder_path, file)

        text = ""

        with pdfplumber.open(pdf_path) as pdf:
            for page in pdf.pages:
                page_text = page.extract_text()

                if page_text:
                    text += page_text + "\n"

        resume_texts[file] = text

print("Total Resumes Read:", len(resume_texts))

for name in resume_texts:
    print(name)
```

```python
#Calculate score for all resumes
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

job_description = """
Looking for candidates with Python, Machine Learning,
SQL, NLP and Data Analysis skills.
"""

results = []

for resume_name, resume_text in resume_texts.items():

    documents = [job_description, resume_text]

    vectorizer = TfidfVectorizer(stop_words="english")

    tfidf_matrix = vectorizer.fit_transform(documents)

    score = cosine_similarity(
        tfidf_matrix[0:1],
        tfidf_matrix[1:2]
    )[0][0]

    results.append((resume_name, score * 100))
print(results)
```

```python
for name, score in results:
    print(f"{name} --> {score:.2f}%")
```

```python
#skill Extraction
skills_db = [
    "python",
    "java",
    "sql",
    "machine learning",
    "deep learning",
    "data analysis",
    "data science",
    "nlp",
    "power bi",
    "excel",
    "aws",
    "azure",
    "cloud computing",
    "html",
    "css",
    "javascript",
    "react",
    "django",
    "flask",
    "git"
]
```

```python
def extract_skills(text):

    text = text.lower()

    found_skills = []

    for skill in skills_db:

        if skill in text:
            found_skills.append(skill)

    return found_skills
```

```python
for resume_name, resume_text in resume_texts.items():

    skills = extract_skills(resume_text)

    print("\n", resume_name)
    print("Skills Found:", skills)
```

```python
#Candidates Ranking
results.sort(key=lambda x: x[1], reverse=True)

for rank, (name, score) in enumerate(results, start=1):

    skills = extract_skills(resume_texts[name])

    print(f"\nRank {rank}")
    print(f"Resume: {name}")
    print(f"Match Score: {score:.2f}%")
    print(f"Skills Found: {', '.join(skills)}")
```

```python
#Candidates Status
results.sort(key=lambda x: x[1], reverse=True)

for rank, (name, score) in enumerate(results, start=1):

    skills = extract_skills(resume_texts[name])

    if score >= 70:
        status = "Shortlisted"
    elif score >= 25:
        status = "Consider"
    else:
        status = "Rejected"

    print(f"\nRank {rank}")
    print(f"Resume: {name}")
    print(f"Match Score: {score:.2f}%")
    print(f"Skills Found: {', '.join(skills)}")
    print(f"Status: {status}")
```

```python
import pandas as pd

data = []

for name, score in results:

    skills = ", ".join(extract_skills(resume_texts[name]))

    if score >= 30:
        status = "Shortlisted"
    elif score >= 15:
        status = "Consider"
    else:
        status = "Rejected"

    data.append([
        name,
        round(score, 2),
        skills,
        status
    ])

df = pd.DataFrame(
    data,
    columns=["Resume", "Match Score (%)", "Skills", "Status"]
)

df.to_csv("final_ranking.csv", index=False)

print(df)
```
